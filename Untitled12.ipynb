```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import chi2
from statsmodels.stats.weightstats import ztest
import statsmodels.api as sm
from scipy.stats import chisquare
from scipy.stats import f_oneway
from scipy.stats import levene
from scipy.stats import gaussian_kde




np.random.seed(123)
```

```python
n1 = 150
n2 = 250
```

```python
sample1 = chi2.rvs(10, size=n1) #X^2(n)
sample2 = np.random.uniform(5, 15, n2) #R(a, b)

```

```python
sampleO = np.concatenate((sample1, sample2), axis=0)
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.hist(sample1, bins=20)
plt.subplot(2,2,2)
plt.hist(sample2, bins=20)
plt.show()
```

```python
# Среднее значение
mean1 = np.mean(sample1)
mean2 = np.mean(sample2)
meanO = np.mean(sampleO)
print(mean1, mean2, meanO)
```

```python
# Вычисление выборочной дисперсии
variance1 = np.var(sample1, ddof=0)
variance2 = np.var(sample2, ddof=0)
varianceO = np.var(sampleO, ddof=0)

print(variance1, variance2, varianceO)
```

```python
# Вычисление с. к. о.
std_deviation1 = np.std(sample1, ddof=0)
std_deviation2 = np.std(sample2, ddof=0)
std_deviationO = np.std(sampleO, ddof=0)
print(std_deviation1, std_deviation2, std_deviationO)
```

Математическое ожидание (среднее значение) распределения хи-квадрат (χ²) равно его числу степеней свободы (df) или, другими словами, математическому ожиданию равномерного распределения.

```python

#Z-Test
# H0: M(X1) = 10
# H1: M(X1) != 10
z_statistic, p_value = ztest(sample1, value=10)
print(f"Z-статистика: {z_statistic}")
print(f"P-значение: {p_value}")
```

```python
#t-Test
# H0: M(X1) = 10
# H1: M(X1) != 10
t_statistic, p_value = stats.ttest_1samp(sample1, 10)
print(f"T-статистика: {t_statistic}")
print(f"P-значение: {p_value}")
```

```python
#X^2-Test izvwstno m = 10
# H0: F(X) = X^2(X)
# H1: F(X) != X^2(X)
statistic, p_value = chisquare(f_obs=np.histogram(sample1, bins=3000)[0], ddof=10)
print(f"статистика: {statistic}")
print(f"P-значение: {p_value}")
```

```python
#X^2-Test izvwstno m = ?
# H0: F(X) = X^2(X)
# H1: F(X) != X^2(X)
statistic, p_value = chisquare(f_obs=np.histogram(sample1, bins=3000)[0])
print(f"Z-статистика: {statistic}")
print(f"P-значение: {p_value}")
```

```python
# 2-sample t-test
# H0: M(F1) == M(F2)
statistic, p_value = stats.ttest_ind(sample1, sample2)

# Выводим результаты
print(f"t-статистика: {statistic}")
print(f"P-значение: {p_value}")
```

```python
# 2-sample F-test (m – изв)
# H0: M(F1) == M(F2)

# Рассчитываем сумму квадратов отклонений для обеих выборок
ss1 = sum((x - 10) ** 2 for x in sample1)
ss2 = sum((x - 10) ** 2 for x in sample2)

# Выполняем двухвыборочный F-тест
F_statistic = (ss1 / (len(sample1) - 1)) / (ss2 / (len(sample2) - 1))


# Вычисление p-значения
p_value = 1 - f_oneway(sample1, sample2).pvalue

print(f"F-статистика: {F_statistic}")
print(f"P-значение: {p_value}")
```

```python
# 2-sample F-test (m – ?)
# H0: M(F1) == M(F2)
F_statistic, p_value = levene(sample1, sample2)

print(f"F-статистика: {F_statistic}")
print(f"P-значение: {p_value}")
```

```python
# Распределение статистики (N(0, 1))
def Z(s1, s2):
    d1 = 20
    d2 = 25/3
    X1_mean = np.mean(s1)
    X2_mean = np.mean(s2)
    zn = ((d1 / n1) + (d2/n2)) ** 0.5
    return (X1_mean - X2_mean) / zn




N = 10000
s1 = chi2.rvs(10, size=n1) #X^2(n)
s2 = np.random.uniform(5, 15, n2) #R(a, b)
valuesZ = []

# Проводим N экспеиментов
for i in range(N):
    s1 = chi2.rvs(10, size=n1) #X^2(n)
    s2 = np.random.uniform(5, 15, n2) #R(a, b)
    valuesZ.append(Z(s1, s2))


# Расчет параметров для Z
M = np.mean(valuesZ)
print("M=", M)

variance = np.var(valuesZ)
print("Дисперсия:", variance)

std_deviation = np.std(valuesZ)
print("Стандартное отклонение:", std_deviation)

# Рисуем гистограмму
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.hist(valuesZ, bins=50)
plt.title('Гистограмма')
plt.show()

# Функция распределения
def plot_cumulative_distribution(data):
    sorted_data = np.sort(data)
    n = len(sorted_data)
    probabilities = np.arange(1, n + 1) / n

    plt.step(sorted_data, probabilities, where='post')
    plt.xlabel('Значения')
    plt.ylabel('Вероятность')
    plt.title('Функция распределения')
    plt.show()

plot_cumulative_distribution(valuesZ)


# плотность распределения
kde = gaussian_kde(valuesZ)

# значения для построения графика
x = np.linspace(min(valuesZ) - 1, max(valuesZ) + 1, 1000)
y = kde(x)

# график плотности распределения
plt.plot(x, y)
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.title('Плотность распределения')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

sorted_data = np.sort(valuesZ)

# Вычислите вероятности для каждого значения в массиве
n = len(sorted_data)
probabilities = np.arange(1, n + 1) / n

# Создайте интерполяцию функции распределения
cdf = interp1d(sorted_data, probabilities, kind='linear')

# значения функции распределения в точках
x_values = sorted_data
y_values = cdf(x_values)
print(x_values) # x
print(y_values) # F(x), где F - функция распределения статистики Z

answer = [] # массив значений p-value

# Создаем массив значений p-value
for i in range(len(x_values)):
    yv = y_values[i]
    answer.append(2 * min(yv, 1 - yv))

# data = y_values
data = answer

# гистограмма
plt.hist(data, bins=30, edgecolor='black')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Гистограмма')
plt.show()

# Функция распределения
def plot_cumulative_distribution(data):
    sorted_data = np.sort(data)
    n = len(sorted_data)
    probabilities = np.arange(1, n + 1) / n

    plt.step(sorted_data, probabilities, where='post')
    plt.xlabel('Значения')
    plt.ylabel('Вероятность')
    plt.title('Функция распределения')
    plt.show()

plot_cumulative_distribution(data)


# плотность распределения
kde = gaussian_kde(data)

# значения для построения графика
x = np.linspace(min(data) - 1, max(data) + 1, 1000)
y = kde(x)

# график плотности распределения
plt.plot(x, y)
plt.xlabel('Значения')
plt.ylabel('Плотность')
plt.title('Плотность распределения')
plt.legend()
plt.show()



# Расчет параметров для P-value
M = np.mean(y_values)
print("M=", M)

variance = np.var(y_values)
print("Дисперсия:", variance)

std_deviation = np.std(y_values)
print("Стандартное отклонение:", std_deviation)
```
