```python

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import chi2
from statsmodels.stats.weightstats import ztest
import statsmodels.api as sm
from scipy.stats import chisquare
from scipy.stats import f_oneway
from scipy.stats import levene
from scipy.stats import gaussian_kde
from scipy.stats import norm, chi2
from scipy.stats import chi2_contingency
from scipy.stats import kstest, norm



np.random.seed(123)
```

```python
#Критерий Хи-квадрат для  H0: X ~ N(m, s)
# параметры распределения
m = 10
s = np.sqrt(20)
nBins = [5, 10, 15, 50]
alpha = 0.05


for B in nBins:
    # Группировка данных по интервалам и подсчет частот
    bins = np.histogram_bin_edges(sample1, bins=B)
    observed, _ = np.histogram(sample1, bins=bins)

    # Ожидаемые частоты для нормального распределения
    expected = len(sample1) * np.diff(norm.cdf(bins, loc=m, scale=s))

    # Вычисляем статистику Хи-квадрат
    chi_squared_stat = np.sum((observed - expected)**2 / expected)

    degrees_of_freedom = len(bins) - 2

    # p-value
    p_value = 1 - chi2.cdf(chi_squared_stat, degrees_of_freedom)
    print("Bins =", B)
    print(f"Хи-квадрат статистика: {chi_squared_stat}")
    print(f"p-value: {p_value}")
    if p_value < alpha:
        print("Отклоняем нулевую гипотезу: распределение не является нормальным.")
    else:
        print("Не отклоняем нулевую гипотезу: распределение является нормальным.")
    print()

```

```python
#Критерий Хи-квадрат для  H0: X ~ R
# параметры распределения
m = 10
s = np.sqrt(20)
nBins = [5, 10, 15, 50]
alpha = 0.05


for B in nBins:
    # Группировка данных по интервалам и подсчет частот
    bins = np.histogram_bin_edges(sample1, bins=B)
    observed, _ = np.histogram(sample1, bins=bins)

    # Ожидаемые частоты для равномерного распределения
    expected = len(sample1) / len(bins)

    # Вычисляем статистику Хи-квадрат
    chi_squared_stat = np.sum((observed - expected)**2 / expected)

    degrees_of_freedom = len(bins) - 2

    # p-value
    p_value = 1 - chi2.cdf(chi_squared_stat, degrees_of_freedom)
    print("Bins =", B)
    print(f"Хи-квадрат статистика: {chi_squared_stat}")
    print(f"p-value: {p_value}")
    if p_value < alpha:
        print("Отклоняем нулевую гипотезу: распределение не является равномерным.")
    else:
        print("Не отклоняем нулевую гипотезу: распределение является равномерным.")
    print()

```

```python
#Критерий Хи-квадрат для  H0: X ~ Хи^2(5)
# параметры распределения
m = 10
s = np.sqrt(20)
nBins = [5, 10, 15, 50]
alpha = 0.05


for B in nBins:
    # Группировка данных по интервалам и подсчет частот
    bins = np.histogram_bin_edges(sample1, bins=B)
    observed, _ = np.histogram(sample1, bins=bins)

    # Ожидаемые частоты для Хи^2(5) распределения
    expected = len(sample1) * np.diff(chi2.cdf(bins, df=5))

    # Вычисляем статистику Хи-квадрат
    chi_squared_stat = np.sum((observed - expected)**2 / expected)

    degrees_of_freedom = len(bins) - 2

    # p-value
    p_value = 1 - chi2.cdf(chi_squared_stat, degrees_of_freedom)
    print("Bins =", B)
    print(f"Хи-квадрат статистика: {chi_squared_stat}")
    print(f"p-value: {p_value}")
    if p_value < alpha:
        print("Отклоняем нулевую гипотезу: распределение не является хи-квадрат распределением с 5 степенями свободы..")
    else:
        print("Не отклоняем нулевую гипотезу: распределение является хи-квадрат распределением с 5 степенями свободы..")
    print()

```

```python
# Chi-squared тест для проверки H0: X ~ Y

from scipy.stats import chi2_contingency

X = sample1
Y = np.random.uniform(5, 15, n1) #R(a, b) Меняем размер Равномерного расперделения для возможности проверки Chi-squared критерия

# Создание категорий на основе квантилей
categories = np.percentile(np.concatenate([X, Y]), [25, 50, 75])

# Создание таблицы сопряженности
table = np.histogram2d(X, Y, bins=[categories, categories])[0]

# Вывод таблицы сопряженности
print("Таблица сопряженности:")
print(table)

# Применение критерия хи-квадрат
chi2_stat, p_value, _, _ = chi2_contingency(table)

print("\nСтатистика хи-квадрат:", chi2_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("\nОтклоняем нулевую гипотезу: Распределения X и Y различны.")
else:
    print("\nНе отклоняем нулевую гипотезу: Распределения X и Y схожи.")
```

```python
# KS-test тест для проверки H0: X ~ Y
from scipy.stats import ks_2samp

X = sample1
Y = sample2

# Применение двухвыборочного KS-теста
ks_statistic, p_value = ks_2samp(X, Y)

print("Статистика KS:", ks_statistic)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("\nОтклоняем нулевую гипотезу: Распределения X и Y различны.")
else:
    print("\nНе отклоняем нулевую гипотезу: Распределения X и Y идентичны.")
```

```python
# Sign test тест для проверки H0: med(X) = med(Y)

from scipy.stats import wilcoxon

X = sample1
Y = np.random.uniform(5, 15, n1) #R(a, b) Меняем размер Равномерного расперделения для возможности проверки Chi-squared критерия

# Применение двухвыборочного Sign теста
statistic, p_value = wilcoxon(X, Y)

print("Статистика Sign теста:", statistic)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("\nОтклоняем нулевую гипотезу: Медианы X и Y различны.")
else:
    print("\nНе отклоняем нулевую гипотезу: Медианы X и Y равны.")
```

```python
# U-test тест для проверки H0: med(X) = med(Y)
from scipy.stats import mannwhitneyu

X = sample1
Y = sample2

# Применение двухвыборочного U-теста
statistic, p_value = mannwhitneyu(X, Y)

print("Статистика U:", statistic)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("\nОтклоняем нулевую гипотезу: Медианы X и Y различны.")
else:
    print("\nНе отклоняем нулевую гипотезу: Медианы X и Y равны.")
```
