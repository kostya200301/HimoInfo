```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import chi2
from statsmodels.stats.weightstats import ztest
import statsmodels.api as sm
from scipy.stats import chisquare
from scipy.stats import f_oneway
from scipy.stats import levene
from scipy.stats import gaussian_kde
from scipy.stats import norm, chi2
from scipy.stats import chi2_contingency
from scipy.stats import kstest, norm



np.random.seed(123)
```

```python
n = 150
```

```python
sample1 = chi2.rvs(10, size=n) #X^2(n)
sample2 = np.random.uniform(5, 15, n) #R(a, b)
```

```python
mean_value = np.mean(sample1)
print("Выборочное среднее:", mean_value)

variance_value = np.var(sample1, ddof=10)
print("Выборочная дисперсия:", variance_value)

standard_deviation = np.std(sample1, ddof=10)
print("Оценка стандартного отклонения:", standard_deviation)
```

```python
mean_value = np.mean(sample2)
print("Выборочное среднее:", mean_value)

variance_value = np.var(sample2)
print("Выборочная дисперсия:", variance_value)

standard_deviation = np.std(sample2)
print("Оценка стандартного отклонения:", standard_deviation)
```

```python
# KK по пирсону
from scipy.stats import pearsonr

x = sample1
y = sample2

# Вычисляем коэффициент корреляции Пирсона и p-значение
correlation_coefficient, p_value = pearsonr(x, y)

print(f"Коэффициент корреляции Пирсона: {correlation_coefficient}")
print(f"P-значение: {p_value}")
```

```python
# KK по Спирмену
from scipy.stats import spearmanr

x = sample1
y = sample2

# Вычисляем коэффициент корреляции Спирмена и p-значение
correlation_coefficient, p_value = spearmanr(x, y)

print(f"Коэффициент корреляции Спирмена: {correlation_coefficient}")
print(f"P-значение: {p_value}")
```

```python
# KK по Кендаллу
from scipy.stats import kendalltau

x = sample1
y = sample2

# Вычисляем коэффициент корреляции Кендалла и p-значение
correlation_coefficient, p_value = kendalltau(x, y)

print(f"Коэффициент корреляции Кендалла: {correlation_coefficient}")
print(f"P-значение: {p_value}")
```

```python
# Строим таблицу сопряженности
matrix_zeros = np.zeros((5, 5))

for i in range(n):
    for i1 in range(len(bins1) - 1):
        for i2 in range(len(bins1) - 1):
            if ((bins1[i1] <= sample1[i] <= bins1[i1 + 1]) and (bins2[i2] <= sample2[i] <= bins2[i2 + 1])):
                matrix_zeros[i1, i2] += 1
print(matrix_zeros)
```

```python
chi2_stat, p_value, dof, expected = chi2_contingency(matrix_zeros)

# Вывод результатов
print(f"Статистика хи-квадрат: {chi2_stat}")
print(f"P-значение: {p_value}")
print(f"Степени свободы: {dof}")
print("Ожидаемые частоты:")
print(expected)
```

```python
def U(L, X, Y):
    return L * X + (1 - L) * Y

def V(L, X, Y):
    return L * X**3 + (1 - L) * Y**3
```

```python
# KK по Спирмену
def KK_by_Spirmen(x, y):
    correlation_coefficient, p_value = spearmanr(x, y)
    return correlation_coefficient

# KK по Кендаллу
def KK_by_Kendall(x, y):
    correlation_coefficient, p_value = kendalltau(x, y)
    return correlation_coefficient

# KK по Пирсону
def KK_by_Pirsony(x, y):
    correlation_coefficient, p_value = pearsonr(x, y)
    return correlation_coefficient
```

```python
# Собираем данные для построения графика зависимости коэффициента корреляции Пирсона(XU),
# коэффициента корреляции по Спирмену(XU), по Кендаллу(XU) от L v [0, 1]
mas_Sprimen = []
mas_Kendall = []
mas_Pirson = []
xx = []
for L in range(0, 5_000):
    l = L / 5_000
    xx.append(l)
    mas_Sprimen.append(KK_by_Spirmen(sample1, U(l, sample1, sample2)))
    mas_Kendall.append(KK_by_Kendall(sample1, U(l, sample1, sample2)))
    mas_Pirson.append(KK_by_Pirsony(sample1, U(l, sample1, sample2)))

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = xx
y1 = mas_Pirson
y2 = mas_Sprimen
y3 = mas_Kendall

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='Pirson', linestyle='-', color='blue')

plt.plot(x, y2, label='Sprimen', linestyle='--', color='green')

plt.plot(x, y3, label='Kendall', linestyle='-.', color='red')

plt.legend()

plt.title('Pirson(XU)(l), Sprimen(XU)(l), Kendall(XU)(l)')
plt.xlabel('L')
plt.ylabel('KK')

# Показать графики
plt.show()
```

```python
# Собираем данные для построения графика зависимости коэффициента корреляции Пирсона(XV),
# коэффициента корреляции по Спирмену(XV), по Кендаллу(XV) от L v [0, 1]
mas_Sprimen2 = []
mas_Kendall2 = []
mas_Pirson2 = []
xx2 = []
for L in range(0, 5_000):
    l = L / 5_000
    xx2.append(l)
    mas_Sprimen2.append(KK_by_Spirmen(sample1, V(l, sample1, sample2)))
    mas_Kendall2.append(KK_by_Kendall(sample1, V(l, sample1, sample2)))
    mas_Pirson2.append(KK_by_Pirsony(sample1, V(l, sample1, sample2)))
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = xx2
y1 = mas_Pirson2
y2 = mas_Sprimen2
y3 = mas_Kendall2

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label='Pirson', linestyle='-', color='blue')

plt.plot(x, y2, label='Sprimen', linestyle='--', color='green')

plt.plot(x, y3, label='Kendall', linestyle='-.', color='red')

plt.legend()

plt.title('Pirson(XV)(l), Sprimen(XV)(l), Kendall(XV)(l)')
plt.xlabel('L')
plt.ylabel('KK')

# Показать графики
plt.show()
```
